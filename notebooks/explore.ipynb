```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# !pip install pandas
```

# Entendimento dos Dados

```python
df = pd.read_excel('../data/DadosProdutividade_2024.xlsx')

# Descrição das Variáveis
print("Descrição das Variáveis:")
print(df.columns)

# Tipos de Dados
print("\nTipos de Dados:")
print(df.dtypes)

# Valores Ausentes
print("\nValores Ausentes por Coluna:")
print(df.isnull().sum())

# Distribuição de Valores Ausentes
print("\nDistribuição de Valores Ausentes (Percentual):")
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)


# Visualização da Distribuição de Valores Ausentes
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_percentage.index, y=missing_percentage.values)
plt.xticks(rotation=90)
plt.title('Percentual de Valores Ausentes por Coluna')
plt.ylabel('Percentual')
plt.xlabel('Colunas')
plt.show()
```

```python
# Contagem de ocorrências por ano
occurrences_by_year = df['ANO_BO'].value_counts().sort_index()

# Contagem de ocorrências por mês
occurrences_by_month = df['MES_ESTATISTICA'].value_counts().sort_index()

print("\nOcorrências por Ano:")
print(occurrences_by_year)

print("\nOcorrências por Mês:")
print(occurrences_by_month)

# Visualização das ocorrências por mês
plt.figure(figsize=(10, 6))
sns.barplot(x=occurrences_by_month.index, y=occurrences_by_month.values)
plt.title('Ocorrências por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Ocorrências')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
